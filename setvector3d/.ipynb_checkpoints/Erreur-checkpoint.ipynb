# Reproduction Erreur

```python
import numpy as np
from sklearn.neighbors import KernelDensity
import matplotlib.pyplot as plt
```

## Load 3d vector

```python
#uniform distribution of vector on a sphere see for explication : http://neilsloane.com/icosahedral.codes/
val=np.loadtxt('icover.3.10832.19.19.txt') 
dim=int(np.size(val)/3)
Vector=val.reshape([dim,3])

theta=np.arctan2(Vector[:,1],Vector[:,0])
phi=np.arccos(Vector[:,2])-np.pi/2
```

```python
kde2 = KernelDensity(bandwidth=0.1, metric='haversine',kernel='gaussian')
kde2.fit(np.transpose(np.array([theta,phi])))
weights=kde2.score_samples(np.transpose(np.array([theta,phi])))
```

```python
X=Vector[:,0]
Y=Vector[:,1]
Z=Vector[:,2]
                
        

id=np.where(Z<0)
X[id]=-X[id]
Y[id]=-Y[id]
Z[id]=-Z[id]
        
        
        
LpL=1./(1.+Z)
xx=LpL*X
yy=LpL*Y
# Prepare the contour plot
plt.figure(figsize=(10,10),dpi=160)
plt.scatter(xx, yy, c=np.exp(weights), s=20, edgecolor='')
plt.colorbar(orientation='vertical',aspect=4,shrink=0.5)


# From here I plot cirle for phi and line for Theta
# compute a 3 circle
for i in [0.,30.,60.]:
    omega = np.linspace(0, 2*np.pi, 1000)
    x_circle = 1./(1.+np.sin(i*np.pi/180.))*np.cos(i*np.pi/180.)*np.cos(omega)
    y_circle = 1./(1.+np.sin(i*np.pi/180.))*np.cos(i*np.pi/180.)*np.sin(omega)
    if i==0:
        plt.plot(x_circle, y_circle,'k', linewidth=3)
    else:
        plt.plot(x_circle, y_circle,'k', linewidth=1.5)
        plt.text(x_circle[200], y_circle[300]+0.04,'$\phi$='+str(90.-i)+'°')
           
# plot Theta line
plt.plot([0,0],[-1,1],'k', linewidth=1.5)
plt.text(1-0.15, 0+0.04,'$\Theta$=0°')
plt.text(-1+0.1, 0-0.06,'$\Theta$=180°')
plt.plot([-1,1],[0,0],'k', linewidth=1.5)
plt.text(-0.16, 1-0.25,'$\Theta$=90°')
plt.text(0.01, -1+0.15,'$\Theta$=270°')
plt.plot([-0.7071,0.7071],[-0.7071,0.7071],'k', linewidth=1.5)
plt.plot([-0.7071,0.7071],[0.7071,-0.7071],'k', linewidth=1.5)
          
            
# draw a cross for x and y direction
plt.plot([1, 0],[0, 1],'+k',markersize=12)
# write axis
plt.text(1.05, 0, r'X')
plt.text(0, 1.05, r'Y')
plt.axis('equal')
plt.axis('off')
```
