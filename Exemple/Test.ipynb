# Test setvector3d object

```python
import setvector3d.setvector3d as sv3d
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity
```

```python
#uniform distribution of vector on a sphere see for explication : http://neilsloane.com/icosahedral.codes/
val=np.loadtxt('icover.3.10832.19.19.txt') 
dim=int(np.size(val)/3)
Vector=val.reshape([dim,3])
```

```python
Vref=np.array([1.,1.,1.])/3**0.5
angle=20.*np.pi/180.
lim=np.cos(angle)

ps=Vector[:,0]*Vref[0]+Vector[:,1]*Vref[1]+Vector[:,2]*Vref[2]
id=np.where(np.abs(ps)>lim)
v=sv3d.setvector3d(Vector[id[0],:])
```

```python
plt.figure(figsize=(8,8))
v.stereoplot(contourf=False,bw=0.1,plotOT=True,nbpoints=0,projz=0,angle=np.array([30.,60.]))
```
